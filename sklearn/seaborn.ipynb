```python
import pandas as pd
data = pd.read_csv('Absenteeism_at_work.csv',sep=';')
data
```

```python
%matplotlib inline
import seaborn as sns
```

# 1 散点图
    jointplot函数

```python
sns.jointplot(x = 'ID',y = 'Absenteeism time in hours',data = data)
```

```python
sns.lmplot(x = 'ID',y = 'Absenteeism time in hours',data = data)
```

```python
sns.lmplot(x = 'ID',y = 'Absenteeism time in hours',hue = 'Seasons',data = data)#通过hue参数制定第二个分组特征
```

# 2 直方图
    distplot函数

```python
sns.distplot(data['Absenteeism time in hours'],bins=40)
```

```python
#使用matplotlib的hist函数绘制直方图
import matplotlib.pyplot as plt
plt.hist(data['Absenteeism time in hours'],bins=50)
```

```python
#distplot函数默认绘制直方图金额KDE（核密度图），不需要核密度图时，kde参数设置为False
sns.distplot(data['Absenteeism time in hours'],bins=40,kde=False)
```

```python
sns.distplot(data['Seasons'],bins=4,kde=False)#观察季节分布
```

# 3 盒图
    boxplot函数

```python
sns.boxplot(data['Absenteeism time in hours'],showfliers=False)#参数orient设置盒图朝向
```

```python
plt.figure(figsize=(10,6))
#columns = data.columns
#columns_new = columns.pop(['ID','Absenteeism time in hours','Month of absence','Day of the week'],axis=1)#保留影响因素
sns.boxplot(x ='ID',y = 'Absenteeism time in hours',data = data,orient="v")
```

```python
plt.figure(figsize=(14,6))
sns.boxplot(x ='ID',y = 'Absenteeism time in hours',hue='Social drinker',data = data,orient="v")#第二个分类变量由hue制定
```

# 4 柱状图
    countplot函数

```python
sns.countplot(x='Absenteeism time in hours',data = data)#绘制每一种取值的样本数量
```

```python
sns.countplot(x='Absenteeism time in hours',hue='Social drinker',data = data)
```

# 5 核密度
    kdeplot函数

```python
sns.kdeplot(data['Absenteeism time in hours'])
```

```python
sns.kdeplot(data['Reason for absence'])
```

```python
#绘制两个特征的核密度图
sns.kdeplot(data['Seasons'],data['Reason for absence'])
```

# 6 小提琴图
    violinplot函数

```python
sns.violinplot(x='Seasons',y='Absenteeism time in hours',data = data)
```

```python
sns.violinplot(x='Seasons',y='Absenteeism time in hours',hue='Social drinker',data = data)
```

```python
#将上图的分组分别绘制在盒图的两边，使用split=True参数设置
sns.violinplot(x='Seasons',y='Absenteeism time in hours',hue='Social drinker',split=True,data = data)
```

# 8 点对图
    pairplot函数表示
+ 同时将多个特征两两之间的散点图等通过一条命令进行绘制

```python
sns.pairplot(data[['Seasons','ID','Absenteeism time in hours','Social drinker']])
```

# 8 热力图
    heatmap函数
+ 只管观察特征之间的相关性强弱

```python
import matplotlib.pyplot as plt

data1 = pd.DataFrame(data)[['ID','Seasons','Absenteeism time in hours','Reason for absence','Transportation expense','Distance from Residence to Work']]
corr = data1.corr()#特征的相关系数矩阵

f,ax = plt.subplots(figsize=(10,10))
cmap = sns.diverging_palette(220,10,as_cmap=True)

sns.heatmap(corr, cmap=cmap, vmax=1.0, linewidths=.3, cbar_kws={"shrink": .5}, ax=ax)
plt.show()
```

# seaborn可视化

```python
import pandas as pd
data = pd.read_csv('stock_index.csv',sep=',')#读入数据
data
```

```python
data.info()#查询数据集信息，是否含有均值
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
plt.style.use({'figure.figsize':(10,8)}) #显示图像，而不用plt.show()
```

```python
#创建空图
fig = plt.figure(figsize = (14,12))
#创建子图
ax1 = fig.add_subplot(2,2,1)
ax2 = fig.add_subplot(2,2,2)
ax3 = fig.add_subplot(2,2,3)

#画子图一，选择颜色，线型，标签，显示图例
#折线图
ax1.plot(data['open'],'k--',label = 'open')
ax1.plot(data['high'],'g+',label='high')
ax1.legend(loc='best')

#子图二，设置绘图种类，以及bin的个数
#直方图
ax2.hist(data['p_change'],bins=50,label='p_change')
ax2.legend(loc='best')

#散点图
ax3.scatter(range(len(data['high'])),data['high'],label='high')
ax3.legend(loc='best')
```

```python
#drawstyle用于选择插值方式
plt.plot(data['high'],drawstyle='steps')
plt.xlim([250,350])
```
